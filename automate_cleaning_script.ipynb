```python
def tanggal(x):
    global tanggal_x

    if isinstance(x, datetime) == True :
        tanggal_x = x.strftime('%m-%d-%Y %H:%M:%S') #dt.to_period('M')
        return tanggal_x
    else :
        return tanggal_x

df_raw[2]=df_raw[2].apply(tanggal)

df_raw
```

```python
df_raw[1] = df_raw[1].astype('str')
type(df_raw[1][3])
```

```python
def w_time(x):
    global w_time_x

    if x == 'nan' :
        return w_time_x
            
    else :
        w_time_x = re.fullmatch(pattern='[\d:]+ to [\d:]+', string=x)
        
        if w_time_x is not None:
            w_time_x = w_time_x.group()
            return w_time_x

        else :
            w_time_x = float('NaN')
            return 'This row would be deleted'
             

df_raw[1]=df_raw[1].apply(w_time)

df_raw
```
